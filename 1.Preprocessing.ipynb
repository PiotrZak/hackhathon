### Zaimplementuj wasn funkcj do normalizacji redniej *mean_normalization(X)*.
$$ X' = \frac{X - X.mean()}{X.max() - X.min()}$$

Wykonaj wykres porwnujcy Twoje customowe scalowanie z zaprezentowanymi rozwizaniami (StandardScaler i MinMaxScaler).

P.S. Przyjmij, e X.shape = (m, 1), gdzie m to liczba prbek.

```python
def mean_normalization(X):
    _X = X.ravel()
    X_fited = (X - X.mean()) / (X.max() - X.min())
    X_prim = X_fited.reshape(-1, 1)
    return X_prim
```

## Zad1 *
### Zaimplementuj obiekt do normalizacji redniej *MeanNormalizer* z funkcjami: *fit_trainsform(X)* oraz *inverse_transform(X')* 

$$X = X'*[X.max() - X.min()]+X.mean()$$

```python
class MeanNormalizer():
    def __init__(self):
        self.__aver = None
        self.__min = None
        self.__max = None
    
    def fit_transform(self, X):
        __X = X.ravel()
        
        self.__aver = __X.mean()
        self.__min = __X.max()
        self.__max = __X.min()
        
        X_fited = (__X - self.__aver) / (self.__max - self.__min)
        X_prim = X_fited.reshape(-1, 1)
        return X_prim
    
    def inverse_transform(self, X_prim):
        __X_prim = X_prim.ravel()
        X_orig = self.__aver + __X_prim * (self.__max - self.__min)
        X = X_orig.reshape(-1, 1)
        return X       
```
